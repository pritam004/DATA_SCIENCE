```
# -*- coding: utf-8 -*-
"""lrf_stochastic.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1LW1bE_UmfkC-0kvpftjZ0e9v_8efyGKL
"""

from google.colab import drive

drive.mount('/content/gdrive')

import numpy as np, math
with open('/content/gdrive/My Drive/datasets/ratings.train') as file:
    array2d = [[float(digit) for digit in line.split()] for line in file]
n=array2d[0]
array2d=array2d[1:]
array2d=np.array(array2d)
train_data=array2d[:50000]
valid_data=array2d[50000:60000]
test_data=array2d[70001:int(n[0])]
#print (valid_data)

maxuser=0
for i in range(len(array2d)):
  if maxuser< array2d[i][0]:
    maxuser=array2d[i][0]
maxuser=int(maxuser)
print (maxuser)
maxitem=0
for i in range(len(array2d)):
  if maxitem< array2d[i][1]:
    maxitem=array2d[i][1]
maxitem=int(maxitem)
print (maxitem)
print(n)

#assigning hyperparameter k

k=16
tolerance=0.9
lam=.00001
def global_avg():
  global train_data
  s=0
  for i in range (len(train_data)):
    s+=train_data[i][2]
  return s/len(train_data)
avg=global_avg()
def error_func(U,V):
  global train_data
  err=0
  for i in train_data:
    new_update=np.dot(U[int(i[0]),:],V[:,int(i[1])])
    #print(new_update)
    #print(MATRIX[int(i[0])][int(i[1])])
    err+=((MATRIX[int(i[0])][int(i[1])]-(new_update))**2)
  return math.sqrt(err/len(train_data))

#input matrix
MATRIX=np.zeros((maxuser+1,maxitem+1))
for i in train_data:
  MATRIX[int(i[0])][int(i[1])]=i[2]

# print(MATRIX)

U=np.random.rand(maxuser+1,k)
V=np.random.rand(k,maxitem+1)

# U, V = (1/k)*np.ones([maxuser+1,k]), np.zeros([k,maxitem+1])

# for i in range(MATRIX.shape[1]):
#   s, c = 0, 0
#   for j in range(MATRIX.shape[0]):
#     if MATRIX[j][i] != 0.0:
#       s += MATRIX[j][i]
#       c += 1
#   m = s/c if c > 0 else 0
#   for j in range(k):
#     V[j][i] += m

U.shape, V.shape

t=1
from sklearn.metrics import mean_squared_error
import random

plot_train, plot_validate = [], []

et=.4
error=error_func(U,V)
# while error> tolerance:
for _ in range(15):
  random.shuffle(train_data)
  for i in range(len(train_data)):
    # eta=et
    eta=et/t
    p=int(train_data[i][0])
    q=int(train_data[i][1])
    update=eta*(MATRIX[p][q]-np.dot(U[p],V[:,q]))
    sign = -1 if update < 0 else 1
    update = (sign * math.sqrt(abs(update/k)))
    for o in range(k):
      U[p][o]+=update*V[o][q]-lam*U[p][o]
    for o in range(k):
      V[o][q]+=update*U[p][o]-lam*V[o][q]
    #U[p]=  U[p] -eta*2*(MATRIX[p][q]-np.dot(U[p],V[q]) 
    t=t+1
  error1=error_func(U,V)
  plot_train.append(error1)

  Y = valid_data[:,2].reshape(-1,1)
  y = []
  for x in valid_data:
    y.append(np.dot(U[int(x[0]),:],V[:,int(x[1])]))
  y = np.array(y).reshape(-1,1)

  error2 = np.sqrt(mean_squared_error(Y,y))
  plot_validate.append(error2)

  print(error1, error2)

import matplotlib.pyplot as plt

plt.plot(plot_train)
plt.plot(plot_validate)
plt.title('model accuracy')
plt.ylabel('error')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
