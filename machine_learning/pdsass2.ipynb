```
import numpy as np
float_formatter = lambda x: "%.3f" % x
np.set_printoptions(formatter={'float_kind':float_formatter})
from sklearn.datasets.samples_generator import make_circles
from sklearn.cluster import SpectralClustering, KMeans
from sklearn.metrics import pairwise_distances
from matplotlib import pyplot as plt
import networkx as nx
import seaborn as sns
import re
sns.set()
```

```
with open('/content/gdrive/My Drive/datasets/pdsasscorrectdata/data-4.csv') as file:
    array2d = [[digit for digit in re.split(',|\n',line)] for line in file]
array2d=np.array(array2d)
#array2d=array2d[:1000]

print(array2d.shape)
matrix=[]
for i in range(len(array2d)):
  temp=[]
  for j in range (512):
    temp.append(float(array2d[i][j]))
  matrix.append(temp)
```

```
matrix=np.array(matrix)
print(matrix.shape)
```

```
W = pairwise_distances(matrix, metric="euclidean")
vectorizer = np.vectorize(lambda x: 1 if x < 5 else 0)
W = np.vectorize(vectorizer)(W)
print(W.shape)
```

```
G = nx.random_graphs.erdos_renyi_graph(1000, 0.5)
W = nx.adjacency_matrix(G)
print(W.todense())
# degree matrix
D = np.diag(np.sum(np.array(W.todense()), axis=1))
print('degree matrix:')
print(D)
# laplacian matrix
L = D - W
print('laplacian matrix:')
print(L)
```

```
e, v = np.linalg.eig(L)
# eigenvalues
print('eigenvalues:')
print(e)
# eigenvectors
print('eigenvectors:')
print(v)
```

```
fig = plt.figure()
ax1 = plt.subplot(121)
plt.plot(e)
ax1.title.set_text('eigenvalues')
i = np.where(e < 10e-6)[0]
ax2 = plt.subplot(122)
plt.plot(v[:, i[0]])
fig.tight_layout()
plt.show()
```

```
i = np.where(e < 0.5)[0]
U = np.array(v[:, 19])
km = KMeans(init='k-means++', n_clusters=19)
km.fit(U)
km.labels_
```

```
sc = SpectralClustering(n_clusters=20, affinity='rbf', random_state=0)
sc_clustering = sc.fit(matrix)
print(sc_clustering.labels_)

#plt.scatter(X[:,0], X[:,1], c=sc_clustering.labels_, cmap='rainbow', alpha=0.7, edgecolors='b')
```

```
def find_max_index(X):
  maxi=-1
  indi=-1
  for i in range (len(X)):
    if maxi<X[i] :
      maxi=X[i]
      indi=i
  return indi
```
